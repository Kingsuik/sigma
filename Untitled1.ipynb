```python
import pandas as pd #pd 행렬을 사용할수있게 하는 패키지(확장팩)
import numpy as np #np 행렬 등 수치 연산을 수행하는 선형대수 라이브러리
import seaborn as sns #시각화 (boxplot)
import matplotlib.pyplot as plt #matplotlib 데이터를 플롯으로 그려주는 라이브러리
import matplotlib.pyplot as pls
#import A as B = A모듈을 새용하고 B라고 부를거임

# 샘플 데이터프레임 생성
data = {'Value': [10, 12, 9, 11, 13, 8, 10, 12, 9, 11]} #변수data에 value값 저장
df = pd.DataFrame(data) #변수df에 행렬로 value값 저장

# 박스 플롯 그리기
plt.figure(figsize=(8, 6)) #가로8 세로6 사이즈입력
sns.boxplot(data=df, y='Value') #sns.boxplot(data,y)
plt.title('Box Plot')
plt.xlabel('Value')

# 평균, 표준편차, CpK 값 계산
mean_value = df['Value'].mean() #df['Value']==> value라는 문자를 포함하지않고 숫자만 가져옴/ 평균값함수 데이터저장값.mean()
std_value = df['Value'].std() #표준편차
USL = 15
LSL = 5
Cpk = min((USL - mean_value) / (3 * std_value), (mean_value - LSL) / (3 * std_value))
print(std_value)
# 평균, 표준편차, CpK 값 플롯에 표시
plt.axhline(mean_value, color='r', linestyle='--', label='Mean')
plt.text(0.05, mean_value + 0.2, f'Mean: {mean_value:.2f}', color='r')
plt.axhline(mean_value + std_value, color='g', linestyle='--', label='Std')
plt.text(0.05, mean_value + std_value + 0.2, f'Std: {std_value:.2f}', color='g')
plt.text(0.05, mean_value + std_value + 0.4, f'Cpk: {Cpk:.2f}', color='b')

# 범례 표시
plt.legend()

#히스토그램
pls.figure(figsize=(8, 6)) #가로8 세로6 사이즈입력
sns.distplot(df['Value'])
pls.title(' histogram')
pls.xlabel('Value')
# 그래프 출력
plt.show()
pls.show()

#왜 그런지 궁금하면 print쳐서 알아보면됨
```
